```python
Suppose a dice is rolled:

Possible values: 1, 2, 3, 4, 5, 6

A single dice roll is uniform, not normal.

Now:

Take samples of 3 dice rolls and compute average:

(1, 2, 6) → Avg = 3
(4, 5, 6) → Avg = 5
(2, 3, 4) → Avg = 3
(1, 1, 6) → Avg = 2.67

If you repeat this thousands of times:

Distribution of these averages becomes approximately normal
```
