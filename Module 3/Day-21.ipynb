# calculating Z-score for boston data

```python
from scipy import stats
```

```python
stats.zscore(df)
```

```python
#to identify outliers using zscore.If Zscore value is greater than3 or greater than -3.That point is an outlier.
# IN simple words vaues should be in the range of 3 to -3 
```

```python
z= np.abs(stats.zscore(df))
z
```

```python
threshold =3
np.where(z>threshold)
```

```python
z[436][11]
```

```python
# Get clean data by removing or filtering the outliers
```

```python
df_no_outlier = df[(z<=3).all(axis=1)]
df_no_outlier
```

#Detecting outliers using IQR 

```python
# Find outliers using IQR Score(i.e. IQR=Q3-Q1) 
```

```python
Q1 = df.quantile(0.25)
Q1
```

```python
Q3 =  df.quantile(0.75)
Q3
```

```python
IQR = Q3- Q1
IQR
```

```python
lower_bound = Q1-1.5*IQR
upper_bound = Q3+1.5*IQR
```

```python
oulier_mask = ((df<lower_bound) | (df>upper_bound))
```

```python
#outliers
```

```python
df[oulier_mask.any(axis=1)]
```

```python
#outlier free data
```

```python
df[~oulier_mask.any(axis=1)]
```
