#File Handling

```python
with open("sample.txt","w") as f:
    f.write("Line 1: Hi how are you \n")
    f.write("Line 2 What can i do for you?\n")
print("Done writing")
```

```python
with open("sample.txt","r") as p:
    data = p.read()
    print(data)
```

```python
#Read line by line
with open ("sample.txt","r") as p:
    for line in p:
        print(line.strip() +" hey am an add-on")
```

```python
#Read lines into a list
with open ("sample.txt","r") as p:
    l = p.readlines()
    print(l)
```

```python
#append
with open ("sample.txt","a") as p:
    p.write("Line 4: New line added")
```
