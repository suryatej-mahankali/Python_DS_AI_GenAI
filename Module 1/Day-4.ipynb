while loops with break

```python
count =0
while True:
    if count ==5:
        break
    print(count, end=" ")
    count += 1
```

Continue 

```python
for i in range(20):
    if i%2 ==0:
        continue
    print(i,end = " ")
```

```python
for i in range(5):
    if i == 3:
        pass
    else:
        print(i,end = " ")
    
```

```python
def sum_of_numbers(n):
    total =0
    for i in range(n+1): #0,1,2,3,..50 (0,51)
        total += i #0,0+1,1+2,3+3,6+4,10+5,15+6,21+7,,,1225+50
    return total
```

```python
sum = sum_of_numbers(50)
sum
```

*args — variable positional arguments

```python
def total(*args):
    print(args)
    return sum(args)
```

```python
total(1,2,3,4,5)
```

**kwargs — variable keyword arguments

```python
def profile(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
profile(name="surya",city="Hyderabad",address="xydgg")
```

```python
def flexible(a,b,*args,**kwargs):
    print(f"a={a},b={b}")
    print(f"args={args}")
    print(f"kwargs={kwargs}")
```

```python
flexible(1,2,5,6,x=10,y=89)
```
