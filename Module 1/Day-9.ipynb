```python
class student:
    def __init__(self,name,age,grade):
        self.name = name
        self.age = age
        self.grade = grade
    def introduction(self):
        return f"I am {self.name} and my age is {self.age} and I got {self.grade} grade"
```

```python
s1 =  student("Surya",31,"A+")
s1.introduction()
```

```python
s1.name
```

```python
class Counter:
    total_count = 0
    def __init__(self,name):
        self.name = name
        self.count = 0
        Counter.total_count += 1
    def increment(self):
        self.count +=1
```

```python
c1 = Counter("page_views")
c1.increment()
```

```python
c1.count
```

```python
c1.increment()
c1.count
```

```python
c2 = Counter("api_calls")
c2.increment()
c2.count
```

```python
Counter.total_count
```

```python
class DateParser:
    date_format ="%Y-%m-%d"
    def __init__(self,date_str):
        self.__date_str = date_str
    
    def display(self):
        return self.__date_str
    
```

```python
date =  DateParser("11-may-2023")
date.display()
```

```python
date.__date_str
```

```python
class Bank:
    def __init__(self, balance):
        self.__balance = balance

b = Bank(1000)
b.__balance = -5000  # ❌ wrong but allowed
```

```python
class Bank:
    def __init__(self, balance):
        self.__balance = balance

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
        else:
            print("Not allowed")
    def get_balance(self):
        return self.__balance
```

```python
b = Bank(1000)
b.withdraw(50)
```

```python
b.get_balance()
```
