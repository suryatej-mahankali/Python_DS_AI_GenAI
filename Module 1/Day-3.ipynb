Conditional Statements

```python
Temp = 35
if Temp>40:
    print("Extreme Heat")
elif Temp>30:
    print("It's hot")
elif Temp>20:
    print("It's Pleasant")

else:
    print("It's Cold")
```

Nested If

```python
age = 20
has_id = False
if age >= 18:  # if True:
    if has_id:
        print("Entry Allowed")
    else:
        print("Bring your ID")
else:
    print("Too Young")
```

```python
#enumerate() => index+value   ["Surya",3,4,5,6,7] ==> [0,1,2,3,4,5,...]

fruits  = ["apple","banana","kiwi"]
for id,fruit in enumerate(fruits):
    print(f"{id}:{fruit}")
```

```python
names = ["Surya","nagraj","akhil","anusha"]
scores = [89,90,91,92]
for name,score in zip(names,scores):
    print(f"{name}:{score}")
```

```python
names = ["Surya","nagraj","akhil","anusha"]
scores = [89,90,91,92]
Grades = ["A","AA","AAA","AAAA"]
for name,score,Grade in zip(names,scores,Grades):
    print(f"{name}:{score}:{Grade}")
```
