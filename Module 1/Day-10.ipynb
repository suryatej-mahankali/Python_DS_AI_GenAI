Inheritance

```python
class Animal:
    def __init__(self,name,species):
        self.name= name
        self.species = species
    def speak(self):
        return f"{self.name} makes a sound"
    def info(self):
        return f"{self.name} is a {self.species}"
```

```python
class Dog(Animal):
    def __init__(self,name,breed):
        super().__init__(name,"Dog")
        self.breed = breed
    def speak(self):
        return f"{self.name} says woof!"  # method overriding
    def fetch(self):
        return f"{self.name} fetches the ball"
```

```python
dog = Dog("Buddy","labrdor")
```

```python
dog.speak()
```

```python
dog.info()
```

```python
dog.fetch()
```

#multiple inhertance

```python
class Flyable:
    def fly(self):
        return "I can fly"
class Swimmable:
    def swim(self):
        return "I can swim"
class Duck(Animal,Flyable,Swimmable):
    def __init__(self,name):
        super().__init__(name,"Duck")
    # def speak(self):
    #     return f"{self.name} says quack!"
```

```python
fly = Flyable()
fly.fly()
```

```python
duck = Duck("Donald")
duck.speak()
```

```python
duck.fly()
```

```python
duck.swim()
```

```python
# MRO — Method Resolution Order
print(Duck.__mro__)
# (Duck, Animal, Flyable, Swimmable, object)
```

#Mutli level inhertance

```python
class animal:
    def eat(self):
        return "Animal Eats"
class dog(animal):
    def bark(self):
        return "dog barks"
class puppy(dog):
    def weep(self):
        return "puppy weeps"
```

```python
p = puppy()
p.eat()
```

```python
p.bark()
```

```python
p.weep()
```

```python
d = dog()
d.weep()
```
