Polymorphism

```python
class circle:
    def __init__(self,radius):
        self.radius = radius
    def area(self):
        return 3.1428 *self.radius**2
class rectange:
    def __init__(self,width,height):
        self.width = width
        self.height =  height
    def area(self):
        return self.width * self.height

class triangle:
    def __init__(self,base,height):
        self.base = base
        self.height =  height
    def area(self):
        return 0.5 * self.base * self.height
```

```python
shapes =[circle(5),rectange(5,6),triangle(4,9)]
```

```python
for shape in shapes:
    print(f"{type(shape).__name__}:area ={shape.area():.2f}")
```

```python
class CSVExporter:
    def export(self,data):
        return   ",".join(str(x) for x in data)
class JSONExporter:
    def export(self,data):
        import json
        return json.dumps(data)
```

```python
def save_report(exporter,data):
    print(exporter.export(data))
```

```python
save_report(CSVExporter(),[1,2,3])
```

```python
save_report(JSONExporter(),[1,2,3])
```

Dunder Method	Enables
__str__(self)	print(obj), str(obj)
__repr__(self)	repr(obj), debugger display
__len__(self)	len(obj)
__eq__(self, other)	obj1 == obj2
__lt__(self, other)	obj1 < obj2, sorting
__add__(self, other)	obj1 + obj2
__iter__(self)	for x in obj
__getitem__(self, key)	obj[key]
