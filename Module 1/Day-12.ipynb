# Abstraction

```python
from abc import ABC,abstractmethod
```

```python
#abstract class
class Vehicle(ABC):
    d=   56
    @abstractmethod
    def start(self):
        print("Abstaract method")
    @abstractmethod
    def sample(self):
        print("Sample")
class Car(Vehicle):
    def start(self):
        print("car starts with a key")
class Bike(Vehicle):
    def start (self):
        print("Bike starts with a kick")
```

```python
c =  Car()
c.d
```

#Decorators

```python
#A decorator is a function that takes another function as an argument 
# and returns a new function that extends or alters its behavior without
#  modifying the original source code
import time
def log_decorator(fun):
    def wrapper():
        print("Function start",time.time())
        fun()
        print("function ended",time.time())
    return wrapper

@log_decorator
def process():
    print("Processing Data")

```

```python
process()
```

@property

```python

class circle:
  def __init__(self,radius):
    self._radius = radius # single "_" --> protected variable,double "__" it is private variable

  @property
  def area(self):
    return 3.14 * self._radius **2

```

```python
c = circle(5)
c.area #no ()
```

```python
c._radius
```

@classmethod - Method that works with a class and not with object

```python
class Student:
    school = "ABC school"

    def __init__(self,name):
        self.name = name
    
    @classmethod
    def change_school(cls,new_name):
        cls.school =  new_name

        
```

```python
Student.change_school("XYZ school")
```

```python
s1 = Student("Surya")
s1.school
```

```python
s2 = Student("tej")
```

```python
s2.school
```

```python
Student.change_school("high school")
```

```python
p1 = Student("raj")
p1.school ="HIGH SCHOOL"
```

```python
p1.school
```

```python
p2 =  Student("pooja")
p2.school
```
