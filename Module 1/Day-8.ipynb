#variable scope

```python
message ="Hello"
def show_message():
    print("Inside function",message)
```

```python
show_message()
print("Outside function",message)
```

```python
value = 10
def sample_fun():
    value = value +5
    print(value)
```

```python
sample_fun()
```

```python
value = 10
def correct_fun():
    global value
    value = value +5
    print("Inside fuc:",value)
```

```python
correct_fun()
print("Outside function ",value)
```

```python
x= 100
def test_score():
    global x
    print(x)
    x =  50
    print("Inisde ",x)

    
```

```python
test_score()
print("Global",x)
```
