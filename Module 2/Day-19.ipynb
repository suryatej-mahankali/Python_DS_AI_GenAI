MATPLOTLIB

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x =np.linspace(0,10,9)
x
```

```python
y = x **2
y
```

```python
plt.plot(x,y,'g')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('X vs Y')
```

```python
#create multiple plots
plt.subplot(2,2,1)
plt.plot(x,y,'r',marker="*")
plt.subplot(2,2,2)
plt.plot(y,x,'b', marker = "^")
plt.subplot(2,2,3)
plt.plot(x,y,"g", marker = "^")
plt.subplot(2,2,4)
plt.plot(y,x,"y", marker = "*")

```
