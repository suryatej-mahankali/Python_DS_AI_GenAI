```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x= np.linspace(0,5,11)
y= x**2
fig=plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])# main axes # left,bottom,width,height
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])#small graph
axes1.plot(x,y,'r')
axes1.set_xlabel('x_main')
axes1.set_ylabel('y_main')

axes2.plot(x,y,'g')
axes2.set_xlabel('x_small')
axes2.set_ylabel('y_small')
```

```python
fig= plt.figure()
ax= fig.add_axes([0,0,1,1])
ax.plot(x,y,'b',label = "x**2")
ax.plot(x,x**3,'g',label = 'x**3')
ax.legend()
```
